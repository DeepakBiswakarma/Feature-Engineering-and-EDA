```python
missing_val = df.isnull().sum()/df.shape[0] *100
missing_val
```

```python
missing_var = missing_val[missing_val > 20].keys()
missing_var
```

### Droping the missing variable which contains null value more than 20%.

```python
df2_drop_missing_var = df.drop(columns = missing_var)
```

```python
df2_drop_missing_var.shape
```

```python
df3_num = df2_drop_missing_var.select_dtypes(include=["int64","float64"])
df3_num.head(6)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_num.isnull())
```

```python
# Getting those rows which are having null values.
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
# Taking out only those variables having null values.
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20, kde_kws={"linewidth":5,"color":"#DC143C"})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={"linewidth":5,"color":"red"}, label="Original")
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={"linewidth":5,"color":"blue"}, label="Mean")
    plt.legend()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False,kde_kws={"linewidth":5,"color":"red"}, label="Original")
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws={"linewidth":5,"color":"blue"}, label="Mean")
    sns.distplot(df5_num_median[var],bins=20,hist=False,kde_kws={"linewidth":5,"color":"black"}, label="Median")
    plt.legend()
```

```python
for i,var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
```

```python
df_concat = pd.concat([df3_num[missing_num_var],df4_num_mean[missing_num_var],df5_num_median[missing_num_var]],axis=1)
```
